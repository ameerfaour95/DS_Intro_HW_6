# Q2:

### Adding a new column called 'No_punct_description'

```python
data["No_punct_description"]=data['description']
```

### Creating a function that remove all characters

```python
def remove_characters(characters):
    des=re.split('[\n , : ; . - _ ]', characters)
    while '' in des:
        des.remove('')
    des=' '.join(str(e) for e in des)
    return str(des)
```

```python
data['No_punct_description'] = data['No_punct_description'].apply(remove_characters)
```

### Creating a function that works for Q3 and Q4 , i have to change only the variable 'looking_for' 

```python
def count_val(n):
    counter=0
    des=n.split(' ')
    for x in des:
        if looking_for == x :
            counter+=1
    return (counter)
```

```python
looking_for='restaurants'
data['has_restaurants'] = data['No_punct_description'].apply(count_val)
```

```python
looking_for='museum'
data['has_museums'] = data['No_punct_description'].apply(count_val)
```

# Q5:

### Creating a function that counts the number of times the words “beach” or “ocean” or “sea” appears in the description 

```python
def count_val(n):
    counter=0
    des=n.split(' ')
    for x in des:
        if 'beach' == x or 'ocean'== x or 'sea'== x :
            counter+=1
    return (counter)
```

```python
data['has_beaches'] = data['No_punct_description'].apply(count_val)
```
